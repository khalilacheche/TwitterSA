```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import os
```

```python
import nltk

from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('omw-1.4')
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    text = text.lower()
    text = word_tokenize(text)
    text = [lemmatizer.lemmatize(word) for word in text if word not in stop_words]
    text = ' '.join(text)
    return text
```

```python
pathv = "./data/"
# load the data, read the txt file and convert it to a csv file
df_negative = pd.read_csv(pathv + "train_neg.txt", sep = "\t", header = None)
df_negative.columns = ["text"]
#
df_negative['text'] = df_negative['text'].apply(lemmatize_text)

df_positive = pd.read_csv(pathv + "train_pos.txt", sep = "\t", header = None)
df_positive.columns = ["text"]
#
df_positive['text'] = df_positive['text'].apply(lemmatize_text)
```

```python
# display df_nagative and its length
print(len(df_negative))
df_negative.head(5)
```

```python
# display Positive and its length
print(len(df_positive))
df_positive.head(5)
```

```python
# count the occurence of each word in the text
from collections import Counter
word_count = Counter()
for text in df_positive["text"].values:
    for word in text.split():
        word_count[word] += 1
df_word_count_positive = pd.DataFrame.from_dict(word_count, orient = "index")
df_word_count_positive.columns = ["count_positive"]
df_word_count = df_word_count_positive.sort_values(by = "count_positive", ascending = False)
```

```python
# count the occurence of each word in the text
word_count = Counter()
for text in df_negative["text"].values:
    for word in text.split():
        word_count[word] += 1
df_word_count_negative = pd.DataFrame.from_dict(word_count, orient = "index")
df_word_count_negative.columns = ["count_negative"]
df_word_count_negative = df_word_count_negative.sort_values(by = "count_negative", ascending = False)
```

```python
# combine the two dataframes with the index as the word
df_word_count = pd.concat([df_word_count_negative, df_word_count_positive], axis = 1)
df_word_count = df_word_count.fillna(0)
```

```python
df_word_count ["diff"]= (df_word_count["count_positive"] - df_word_count["count_negative"])
df_word_count = df_word_count.sort_values(by = "diff", ascending = False)
df_word_count.head(30)
```
